```python
dicts = df[['PULocationID', 'DOLocationID']].to_dict(orient='records')
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
dv = DictVectorizer()
X = dv.fit_transform(dicts)
y = df_jan23_no_outliers['duration'].values
```

```python
# Train model
model = LinearRegression()
model.fit(X, y)
```

```python
# Predict & compute RMSE
y_pred = model.predict(X)
rmse = np.sqrt(mean_squared_error(y, y_pred))
print(f'RMSE: {rmse:.2f}')
```

## Feb data

```python
df_feb23_no_outliers['PULocationID'] = df_feb23_no_outliers['PULocationID'].astype(str)
df_feb23_no_outliers['DOLocationID'] = df_feb23_no_outliers['DOLocationID'].astype(str)
val_dicts = df_feb23_no_outliers[['PULocationID', 'DOLocationID']].to_dict(orient='records')
X_val = dv.transform(val_dicts)  # Use the same `dv` from training
```

```python
y_val = df_feb23_no_outliers['duration'].values
y_pred_val = model.predict(X_val)

rmse_val = np.sqrt(mean_squared_error(y_val, y_pred_val))
print(f'Validation RMSE: {rmse_val:.2f}')
```
